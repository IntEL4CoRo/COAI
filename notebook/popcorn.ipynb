# Making Popcorn



## Select a robot
Open a new Launcher (Ctrl-Shift-L), start a terminal and execute the launch file to establish the environment and robot.
```bash
roslaunch cram_pr2_popcorn_demo sandbox.launch
```

## Run the popcorn making task

1. Load the code for the task

````common-lisp
(uiop:run-program 
 "roslaunch cram_pr2_popcorn_demo sandbox.launch &" 
:output t
:error-output t)

(uiop:run-program 
 "roslaunch rvizweb update_config.launch config_file:=${REPO_DIR}/binder/rviz_config/popcorn_pr2_demo.json" 
:output t
:error-output t)

(uiop:run-program 
 "rostopic pub /tf tf2_msgs/TFMessage --file ${REPO_DIR}/binder/rviz_config/popcorn_pr2_demo.txt --once" 
:output t
:error-output t)
````

````common-lisp
(asdf:load-system :cram-pr2-popcorn-demo)
````

2. Initialize the plan executive

````common-lisp
(setf cram-tf:*tf-broadcasting-enabled* t)
(setf cram-tf:*tf-broadcasting-topic* "tf")
(setf cram-tf:*broadcast-environment-from-projection* t)
(roslisp-utilities:startup-ros)
````

3. Start the task

````common-lisp
(urdf-proj:with-simulated-robot (demo::demo))
````

To stop the task, hit the black square in the top bar.
